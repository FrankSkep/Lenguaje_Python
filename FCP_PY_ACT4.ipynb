## Funciones para validar entrada de numeros

```python
# Verifica que se ingrese un valor numerico, si no se proporciona tipo, sera int
def getNum(msg, tipo = int):
    while True:
      try:
        return tipo(input(msg))
      except ValueError:
        print(f"Error: Ingresa un numero {tipo.__name__} valido.")

# Verifica que se ingrese un valor numerico de cualquier tipo en un rango
def getNumInRange(msg, ri, rf, tipo = int):
    while True:
        try:
            num = tipo(input(msg))
            if num >= ri and num <= rf:
                return num
            else:
                print("Error: Ingresa un numero dentro del rango.")
        except ValueError:
            print(f"Error: Ingresa un numero {tipo.__name__} valido.")
```

## Ejercicio 1
Programa en Python que genere 40 números aleatorios entre el 0 y 200, desplegar los números y la leyenda de cada número si es par o impar , la cantidad de los números pares e impares así como la suma de los números pares o impares.

```python
import random

N = 40
cont_p, cont_imp = 0, 0
suma_p, suma_imp = 0, 0

for i in range(N):
    num = random.randint(0, 200)
    if num % 2 == 0:
        print(f"{num} - Par")
        cont_p += 1
        suma_p += num
    else:
        print(f"{num} - Impar")
        cont_imp += 1
        suma_imp += num
print(f"Cantidad pares: {cont_p}\nCantidad impares: {cont_imp}")
print(f"Suma pares: {suma_p}\nSuma impares: {suma_imp}")
```

## Ejercicio 4
Programa en Python que lea n cantidad de números enteros dentro de un rango dado (> 0 ) , el programa deberá terminar cuando el usuario introduzca el número cero. desplegar la suma de números y la media.

```python
ri = getNum("> Ingresa rango inicial (> 0): ")
rf = getNum("> Ingresa rango final (> 0): ")

suma, contador = 0, 0
media = None

while True:
  try:
    num = int(input("> Ingresa un numero (o 0 para salir): "))
    if num >= ri and num <= rf:
        suma += num
        contador += 1
    elif num == 0:
        break
    else:
        print("Error: Ingresa un numero dentro del rango.")
  except ValueError:
        print(f"Error: Ingresa un numero valido.")

if contador > 0:
  media = suma / contador
  print(f"La suma de los {contador} numeros es: {suma}")
  print(f"La media de los numeros es: {media}")
```

## Ejercicio 6
Función que lea n cantidad de números hasta que el usuario lo desee, desplegar la suma de los números, media y valor de los números mayores y menores.

```python
def leer_n_numeros():
    cont, suma = 0, 0

    menor = float('inf') # Valor infinito
    mayor = float('-inf') # Valor -infinito

    continuar = 1 # Bandera que controla el ciclo

    while continuar == 1:
        num = getNum("> Ingresa un numero: ")

        if num > mayor:
            mayor = num
        if num < menor:
            menor = num
        cont += 1
        suma += num

        continuar = getNumInRange("¿Continuar ingresando? (1:Si - 2:No): ", 1, 2)
    return cont, suma, menor, mayor

cont, suma, menor, mayor,  = leer_n_numeros()
print(f"Numeros ingresados: {cont}")
print(f"Suma: {suma}\nMedia: {suma / cont:.2f}")
print(f"Numero menor: {menor}\nNumero mayor: {mayor}")
```

# Ejercicio 7
Función que genere 15 números impares entre 10 y 60 o máximo de 25 números desplegar la media de los pares y media de impares.

```python
import random

def generar_impares():
    cont_pares, cont_impares = 0, 0
    suma_pares, suma_impares = 0, 0

    while cont_impares != 15 and cont_pares + cont_impares < 25:
        num = random.randint(10, 60)
        if num % 2 == 0:
            cont_pares += 1
            suma_pares += num
        else:
            cont_impares += 1
            suma_impares += num

    print(f"Total pares: {cont_pares}")
    print(f"Total impares: {cont_impares}")
    print(f"Total numeros: {cont_pares + cont_impares}")
    return (suma_pares / cont_pares, suma_impares / cont_impares)

media_pares, media_impares = generar_impares()
print(f"Media pares: {media_pares:.2f}")
print(f"Media impares: {media_impares:.2f}")
```

## Ejercicio 8
Función que sirva para leer y validar un número dentro de un rango dado por el usuario. repetir esta acción hasta que el usuario lo desee, desplegar cantidad de números y promedio de los números.

```python
def leer_nums_validos(ri, rf):
    cont, suma = 0, 0
    continuar = 1

    while continuar == 1:
        num = getNumInRange(f"> Ingresa un numero entre {ri} y {rf}: ", ri, rf)
        cont += 1
        suma += num

        continuar = getNumInRange("¿Continuar ingresando? (1:Si - 2:No): ", 1, 2)
    return cont, suma / cont

ri = getNum("> Ingresa rango inicial: ")
rf = getNum("> Ingresa rango final: ")

total_numeros, media = leer_nums_validos(ri, rf)
print(f"El total de numeros ingresados es: {total_numeros}")
print(f"El promedio de los numeros es: {media}")
```
